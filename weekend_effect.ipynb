```python
# 주제 : 주가의 주말효과(weekend effect)
# 가설 : 주말효과가 존재한다면 월요일의 평균 수익률이 낮은 기업들의 수가 다른 요일에 비해 많을 것이다   
# 주말효과란? 월요일 수익률이 다른 요일 수익률에 비해 유난히 낮은 현상
# 기업들은 불리한 정보가 있을 때 주말 저녁 폐장 이후에 발표 -> 월요일에 매도물량이 많아 수익률이 낮아짐 
```

```python
trans_stock = stock.T
```

```python
trans_stock.head()
```

```python
df_stock = 100 * trans_stock.pct_change()
df_stock
```

```python
#인덱스 날짜로 변환 
df_stock.index = pd.to_datetime(df_stock.index)
```

```python
#날짜 -> 요일로 변환 / 요일별 평균 구하기  
df_weekday = df_stock.groupby(df_stock.index.dayofweek).mean()
df_weekday.index = ["Mon", "Tue", "Wed", "Thu", "Fri"]
```

```python
df_weekday
```

```python
weekdays = ["Mon", "Tue", "Wed", "Thu", "Fri"]
kr_weekdays = ["월요일", "화요일", "수요일", "목요일", "금요일"]

for i, day in enumerate(weekdays):
    print(f"{kr_weekdays[i-1]} 대비 {kr_weekdays[i]}의 평균 수익률이 0 이하인 기업의 수 : " , len(df_weekday.loc[day][df_weekday.loc[day] <= 0]))
```

```python
# 월요일의 평균 수익률이 0보다 낮은 기업들이 1761개(총 2274개)로 다른 요일에 비해 많은 것을 알 수 있음 
# 단, 화요일이나 수요일과의 차이는 명확하나 목요일에도 평균 수익률이 낮은 기업들이 많음 
```

```python
# 요일별 기술통계량 확인
describe_df = pd.DataFrame()

for i, day in enumerate(weekdays):
    describe_df[day] = df_weekday.loc[day].describe()

describe_df
```

```python
# 월요일의 수익률 평균과, 목요일의 수익률 평균이 비슷함을 알 수 있음 
# 중앙값 또한 크게 차이나지 않음
# 그렇다면, 주말효과는 없다고 볼 수 있을까?

# 데이터가 2018년~2020년 데이터 이므로 연도별로 확인하기로 함 
# 코로나19 이후로 급락 / 급등이 잦았으므로 연도별 차이가 있을 것으로 판단 
```

```python
years = [2018, 2019, 2020]
weekdays = ["Mon", "Tue", "Wed", "Thu", "Fri"]
kr_weekdays = ["월요일", "화요일", "수요일", "목요일", "금요일"]

for year in years: 
    df_year = df_stock[df_stock.index.year == year]
    df_weekday = df_year.groupby(df_year.index.dayofweek).mean()
    df_weekday.index = ["Mon", "Tue", "Wed", "Thu", "Fri"]
    print("--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--")
    print(f"{year}년도 데이터입니다")
    print("--------")
    for i, day in enumerate(weekdays):
        
        print(f"{kr_weekdays[i-1]} 대비 {kr_weekdays[i]}의 평균 수익률이 0 이하인 기업의 수 : " , len(df_weekday.loc[day][df_weekday.loc[day] <= 0]))
```

```python
# 2018년에는 월요일의 수익률이 0 이하인 기업의 수가 다른 요일에 비해 가장 많았음 
# 하지만 2019년 부터는 월요일의 수익률이 0 이하인 기업의 수보다 목요일의 수익률이 0 이하인 기업이 더 많음 
# 2020년에는 목요일의 수익률이 0 이하인 기업이 뚜렷하게 더 많은 것을 확인할 수 있음 
```

```python
years = [2018, 2019, 2020]
weekdays = ["Mon", "Tue", "Wed", "Thu", "Fri"]
kr_weekdays = ["월요일", "화요일", "수요일", "목요일", "금요일"]

for year in years: 
    df_year = df_stock[df_stock.index.year == year]
    df_weekday = df_year.groupby(df_year.index.dayofweek).mean()
    df_weekday.index = ["Mon", "Tue", "Wed", "Thu", "Fri"]
    print("--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--*--")
    print(f"{year}년도 데이터입니다")
    print("--------")
    
    describe_df = pd.DataFrame()

    for i, day in enumerate(weekdays):
        describe_df[day] = df_weekday.loc[day].describe()

    print(describe_df)
```

```python

# 연도/요일별 기술통계량으로 봤을 때에도, 
# 2019년 부터는 목요일의 평균 수익률이 가장 낮고, 중앙값 또한 가장 낮음 

# 이에 2018년에는 주말효과가 발견되지만, 2019년과 2020년에는 발견할 수 없었다는 것으로 결론을 내려볼 수 있음 
```
